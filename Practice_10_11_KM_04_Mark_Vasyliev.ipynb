### **Завдання 1 (2 бали)**

Розрахуйте коефієнти розкладу біному Ньютона за допомогою генератора (https://uk.wikipedia.org/wiki/%D0%91%D1%96%D0%BD%D0%BE%D0%BC_%D0%9D%D1%8C%D1%8E%D1%82%D0%BE%D0%BD%D0%B0).

1. Степінь многочлена користувач має вводити з клавіатури (невід'ємне ціле число).

2. Програма має підраховувати біноміальні коефіцієнти для біномів усіх степенів, починаючи від 0 і закінчуючи тим, який ввів користувач.

3. Коефіцієнти мають виводитись у вигляді трикутника Паскаля.

Наприклад, якщо користувач введе число 5, програма має вивести результат:

1

1 1

1 2 1

1 3 3 1

1 4 6 4 1

1 5 10 10 5 1


```
def pascal(n):
    raw = [1]
    for i in range(n+1):
        yield raw
        raw = [i+j for i,j in zip(raw + [0], [0] + raw)]

power = int(input("Enter the power: "))
for r in pascal(power):
    print(*r)
```

Декоратор повинен створювати список усіх сторінок у потрібній послідовності, причому номери сторінок кожного зошита мають зберігатись в окремому вкладеному списку. 6. Замість функції з п. 4 створіть генератор, який буде визначати послідовність книжкових сторінок для кожного зошита. Ці рядки можна потім вивести або у циклі, або за допомогою функції next(). Функція-декоратор має працювати з цим генератором так само, як і зі звичайною функцією.

```
def add_mode(active):
    def wrap(func):
        def wrapper(*args, **kwargs):
            if active:
                return list(func(*args, **kwargs))
            return func(*args, **kwargs)
        return wrapper
    return wrap

class NumericError(Exception):
    pass

try:
    book_pages = int(input("Enter the number of pages in the book: "))
    if book_pages < 1 or book_pages > 1280:
        raise NumericError("Page number must be from 1 to 1280!")

    section_pages = int(input("Choose the number of pages in a section (16, 24 or 32): "))
    if section_pages not in (16, 24, 32):
        raise NumericError("Value must be 16, 24 or 32!")
    if book_pages % section_pages != 0:
        raise NumericError("This number must be a factor of the first one!")

    mode = int(input("Choose the mode (0 or 1): "))
    if mode not in (0, 1):
        raise NumericError("Only 0 and 1 are allowed!")
except ValueError:
    print("Value must be an integer!")
except NumericError as ne:
    print(ne)
else:

    @add_mode(mode)
    def page_sequence(bp, sp):
        for k in range(bp // sp):
            lst = [pair[::(-1)**p] for p,pair in enumerate(zip(range((k+1)*sp, k*sp + sp//2, -1),
                                                            range(1 + k*sp, k*sp + sp//2 + 1)))]
            yield [page for pair in lst for page in pair]

    print("page_sequence object:", type(page_sequence(book_pages, section_pages)))
    print(f"Pages in book: {book_pages}\nPages in section: {section_pages}\n"
          f"Number of sections: {book_pages // section_pages}")
    for i,seq in enumerate(page_sequence(book_pages, section_pages)):
        print(i+1, ":", *seq)
```
